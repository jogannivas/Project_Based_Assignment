# Starting EDA and Feature selection

First operation in Review column

```python
df["Reviews"].unique()
```

```python
df[~df["Reviews"].str.isnumeric()]
```

```python
df_cp=df.copy()
```

I have detected a value with 3.0M so here i am removing that row which is in index number 10472 .

```python
df_cp=df_cp.drop(df_cp.index[10472])
```

Converting the column type from object to integer

```python
df_cp["Reviews"]=df_cp["Reviews"].astype("int")
```

```python
df_cp["Reviews"]
```

```python
df_cp.info()
```

Second operation in size column

```python
df_cp["Size"]=df_cp["Size"].str.replace("M","000")
df_cp["Size"]=df_cp["Size"].str.replace("k","")
df_cp["Size"]=df_cp["Size"].replace('Varies with device',np.nan)
df_cp["Size"]=df_cp["Size"].astype("float")
```

Third operation to remove "$","+","," from two column at a time by python code which are "Installs" & "Price" columns.

```python
to_remove=["$","+",","]
from_remove=["Installs" , "Price"]
for item in to_remove:
  for col in from_remove:
    df_cp[col]=df_cp[col].str.replace(item,"")
```

```python
df_cp["Price"]=df_cp["Price"].astype("float")
```

```python
df_cp["Installs"]=df_cp["Installs"].astype("int")
```

Fourthly i am going to convert lastupdate column in to dataand time data type  .

```python
df_cp['Last Updated']=pd.to_datetime(df_cp['Last Updated'])
df_cp['Day'] =df_cp['Last Updated'].dt.day
df_cp['Month'] =df_cp['Last Updated'].dt.month
df_cp['Year'] =df_cp['Last Updated'].dt.year
```

# Let's move twords analysis part and explore it more

```python
df_cp=df_cp.drop_duplicates(subset=["App"],keep="first")
```

Seperating the columns which are coming under numerical dataype

```python
numerical=[data for data in df_cp.columns if df_cp[data].dtype != "O"]
categotical=[tata for tata in df_cp.columns if df_cp[tata].dtype == "O"]
```

Seperating the data which are non numeric

```python
numerical
```

```python
categotical
```

```python
print (" So we have {} elements in numerical {}".format (len(numerical),numerical))
print (" So we have {} elements in categotical {}".format (len(categotical),categotical))
```

```python
plt.figure(figsize=(15,20))
for k in range(1 , len(numerical)):
  plt.subplot(5, 4, k)
  sns.kdeplot(x=df_cp[numerical[k]])
  plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(16,9))
for l in range (1 , len(categotical)):
  plt.subplot(2,3,l)
  sns.countplot(x=df_cp[categotical[l]])
  plt.xticks(rotation=90)
```
